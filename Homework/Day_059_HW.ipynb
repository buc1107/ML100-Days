### 用 digits (3個數字即可 , dataset.load_digits(n_class=3)) 資料嘗試跑 PCA


```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import datasets
from sklearn import decomposition
np.random.seed(5)
%matplotlib inline
digits = datasets.load_digits(n_class=3)
X = digits.data
y = digits.target
```

載入 相關套件 並 執行 PCA 實驗 ...

```python
digits = datasets.load_digits(n_class=3)
pca = decomposition.PCA(n_components=3)
pca.fit(X)
X_t = pca.transform(X)
print("X=",X)
print("X_t=",X_t)
print(y)
```

```python
fig = plt.figure(figsize=(16,10))
ax = Axes3D(fig,rect=[0,0,.95,1],elev=48,azim=45)
for _y in range(3):
    ax.text3D(X_t[y == _y,0].mean(),X_t[y == _y,1].mean(),X_t[y == _y,2].mean()+2,"label="+str(_y),fontsize=20)
y_t = np.choose(y,[1,2,0]).astype(float)
ax.scatter(X_t[:,0],X_t[:,1],X_t[:,2],c=y_t,cmap=plt.cm.nipy_spectral)
```
