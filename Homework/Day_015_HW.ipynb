```python
# Import 需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns # 另一個繪圖-樣式套件
plt.style.use('ggplot')

%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```python
matrix = (2*np.random.random((10,10)))-1
matrix_corr = pd.DataFrame(matrix).corr()
print (matrix_corr)
```

```python
"""
Your Code Here
"""
matrix = (2*np.random.random((10,10)))-1
matrix_corr = pd.DataFrame(matrix).corr()
plt.figure(figsize=(10,10))
"""
Your Code Here
"""
heatmap = sns.heatmap(matrix_corr, cmap = plt.cm.RdYlBu_r, vmin = -0.25, annot = True, vmax = 0.6)

plt.show()
```

```python
nrow = 1000
ncol = 3
"""
Your Code Here
"""
matrix = (2*np.random.random((nrow,ncol)))-1

indice = np.random.choice([0,1,2], size=nrow)
plot_data = pd.DataFrame(matrix, indice)


grid = sns.PairGrid(data = plot_data, size = 3, diag_sharey=False)
"""
Your Code Here
Please replace "..." to correct plot function
"""
grid.map_upper(plt.scatter, alpha = 0.2)
grid.map_diag(sns.kdeplot)
grid.map_lower(sns.kdeplot, cmap = plt.cm.OrRd_r)

plt.show()
```
