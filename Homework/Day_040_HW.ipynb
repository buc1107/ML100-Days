## 練習時間
試著使用 sklearn datasets 的其他資料集 (boston, ...)，來訓練自己的線性迴歸模型，並加上適當的正則話來觀察訓練情形。

```python
from sklearn import datasets, linear_model
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, mean_squared_error
import pandas as pd
bc = datasets.load_breast_cancer()
```

```python
lr = linear_model.LogisticRegression()
X = bc.data
y = bc.target
bc_pd = pd.DataFrame(data=bc.data,columns=bc.feature_names)
bc_pd.head(5)
```

```python
bc_pd2 = (bc_pd - bc_pd.mean())/bc_pd.std()
bc_pd2.head()
```

## LASSO

```python
for _alpha in [1e-5,1e-4,0.001,0.01,0.1,1,5,10]:
    lso = linear_model.Lasso(alpha=_alpha)
    X = bc_pd2.values
    y = bc.target
    X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=48)
    lso.fit(X_train,y_train)
    y_test_pred = lso.predict(X_test)
    #print(y_test_pred)
    ascore = accuracy_score(y_test_pred>0.5,y_test)
    print("-"*30+"alpha = "+str(_alpha)+"-"*30)
    print("coefficient: ",lso.coef_)
    print("accuracy: ", ascore)
    print("\n")
```

```python
for _alpha in [1e-5,1e-4,0.001,0.01,0.1,1,5,10]:
    rid = linear_model.Ridge(alpha=_alpha)
    X = bc_pd2.values
    y = bc.target
    X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=48)
    rid.fit(X_train,y_train)
    y_test_pred = rid.predict(X_test)
    ascore = accuracy_score(y_test_pred>0.5,y_test)
    print("-"*30+"alpha = "+str(_alpha)+"-"*30)
    print("coefficient: ",rid.coef_)
    print("accuracy: ", ascore)
    print("\n")
```
