### 請根據不同的 HOUSETYPE_MODE 對 AMT_CREDIT 繪製 Histogram

```python
# Import 需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns # 另一個繪圖-樣式套件

%matplotlib inline
plt.style.use('ggplot')
import warnings
warnings.filterwarnings('ignore')

# 設定 data_path
dir_data = '../data/'
```

```python
f_app = os.path.join(dir_data, 'application_train.csv')
print('Path of read in data: %s' % (f_app))
app_train = pd.read_csv(f_app)
app_train.head()
```

```python
"""
Your Code Here
"""
unique_house_type = app_train['HOUSETYPE_MODE'].unique()
print(unique_house_type)
nrows = len(unique_house_type) //2 + len(unique_house_type)%2
ncols = 2
print("(nrows,ncols) = (",nrows,ncols,")")
plt.figure(figsize=(10,10))
for i in range(len(unique_house_type)):
    plt.subplot(nrows, ncols, i+1)
    """
    Your Code Here
    """
    app_train.loc[app_train['HOUSETYPE_MODE'] == unique_house_type[i], 'AMT_CREDIT'].hist(bar=20)
    
    plt.title(str(unique_house_type[i]))
plt.show()    
```
