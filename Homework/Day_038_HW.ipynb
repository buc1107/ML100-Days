```python
from sklearn import datasets
```

```python
bc = datasets.load_breast_cancer()
print(dir(bc))
print(bc.DESCR)
```

```python
print(bc.target)
print(bc.target_names)
print(bc.data.shape)
print(bc.target.shape)
```

```python
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
lr = linear_model.LogisticRegression()
X = bc.data
y = bc.target
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=48)
print("X_train.shape = ",X_train.shape)
print("X_test.shape = ",X_test.shape)
print("y_train.shape = ",y_train.shape)
print("y_test.shape = ",y_test.shape)
```

```python
lr.fit(X_train,y_train)
y_test_pred = lr.predict(X_test)
print(y_test)
print(y_test_pred)
ascore = accuracy_score(y_test_pred,y_test)
print("accuracy: ", ascore)
```

```python
y_test_prob = lr.predict_proba(X_test)
print(y_test_prob)
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
bc_pd = pd.DataFrame(data=bc.data,columns=bc.feature_names)
bc_pd.head(5)
```

```python
bc_pd2 = bc_pd.copy()
for feature in bc_pd:
    #print("feature: ", bc_pd[feature])
    sc = StandardScaler()
    sc.fit(bc_pd[feature])
    bc_pd2[feature] = sc.transform(bc_pd[feature])
```

```python
bc_pd.head()
```

```python
bc_pd2.head()
```

```python
lr2 = linear_model.LogisticRegression()
X = bc_pd2.values
y = bc.target
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=48)
print("X_train.shape = ",X_train.shape)
print("X_test.shape = ",X_test.shape)
print("y_train.shape = ",y_train.shape)
print("y_test.shape = ",y_test.shape)
```

```python
lr.fit(X_train,y_train)
y_test_pred = lr.predict(X_test)
print(y_test)
print(y_test_pred)
ascore = accuracy_score(y_test_pred,y_test)
print("accuracy: ", ascore)
```
