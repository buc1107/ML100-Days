### kdeplot(核密度估計圖)


核密度估計(kernel density estimation)是在概率論中用來估計未知的密度函數，屬於非參數檢驗方法之一。通過核密度估計圖可以比較直觀的看出數據本身的分佈特徵。具體用法如下：

seaborn.kdeplot(data,data2=None,shade=False,vertical=False,kernel='gau',bw='scott',gridsize=100,cut=3,clip=None, \
  legend=True,cumulative=False,shade_lowest=True,cbar=False, cbar_ax=None, cbar_kws=None, ax=None, **kwargs)

```python
%matplotlib inline
import numpy as np
import seaborn as sns
sns.set() # seaborn的default設定
```

```python
x = np.random.randn(100)
sns.kdeplot(x)
```

```python
num_list = [100,1000,10000]
for num in num_list:
    x = np.random.randn(num)
    sns.kdeplot(x,label=str(num))
```
